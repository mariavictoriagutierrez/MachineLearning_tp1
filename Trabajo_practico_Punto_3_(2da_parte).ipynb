```r
library(dplyr)

# Leer el archivo Excel
ruta_archivo <- "archivo_combinado1.xlsx"
datos_excel <- read_excel(ruta_archivo)

# Reemplazar celdas vacías con missing
datos_excel <- datos_excel %>%
  mutate_all(~ ifelse(is.na(.), "missing", .))

# Especificar la ruta para el nuevo archivo Excel
nueva_ruta_archivo <- "archivo_combinado2.xlsx"

# Exportar el DataFrame modificado a un nuevo archivo Excel
write.xlsx(datos_excel, nueva_ruta_archivo, rowNames = TRUE)

ruta_archivo <- "archivo_combinado2.xlsx"
datos_excel_tp <- read_excel(ruta_archivo)

datos_filtrados <- datos_excel_tp %>%
  filter(ocu == 1, age >= 18)

datos_filtrados2 <- datos_filtrados %>%
  filter(!(y_salary_m_hu == 0 | y_salary_m_hu == "missing"))


# Especificar la ruta para el nuevo archivo Excel
nueva_ruta_archivo <- "archivo_filtrado3.xlsx"

# Exportar el DataFrame filtrado a un nuevo archivo Excel
write.xlsx(datos_filtrados2, nueva_ruta_archivo, rowNames = TRUE)


datos_filtrados2 <- datos_filtrados2 %>%
  mutate(log_y_salary_m_hu = log(as.numeric(y_salary_m_hu)))

#Controlo que no haya missings
datos_filtrados2 <- datos_filtrados2 %>%
  filter(!is.na(log_y_salary_m_hu))

#Creo edad al cuadrado
datos_filtrados2 <- datos_filtrados2 %>%
  mutate(age_cuadrado = age^2)

#Recodifico maxEducLevel. Los 9 (NS/NR) ahora toman el valor de 1 (None)

datos_filtrados2 <- datos_filtrados2 %>%
  mutate(maxEducLevel = ifelse(maxEducLevel == 9, 1, maxEducLevel))

#Creo 7 variables binarias de educación
datos_filtrados2 <- datos_filtrados2 %>%
  mutate(
    educ1 = as.numeric(maxEducLevel == 1),
    educ2 = as.numeric(maxEducLevel == 2),
    educ3 = as.numeric(maxEducLevel == 3),
    educ4 = as.numeric(maxEducLevel == 4),
    educ5 = as.numeric(maxEducLevel == 5),
    educ6 = as.numeric(maxEducLevel == 6),
    educ7 = as.numeric(maxEducLevel == 7)
  )

#Creo 9 variables binarias de relación laboral
datos_filtrados2 <- datos_filtrados2 %>%
  mutate(
    relab1 = as.numeric(relab == 1),
    relab2 = as.numeric(relab == 2),
    relab3 = as.numeric(relab == 3),
    relab4 = as.numeric(relab == 4),
    relab5 = as.numeric(relab == 5),
    relab6 = as.numeric(relab == 6),
    relab7 = as.numeric(relab == 7),
    relab8 = as.numeric(relab == 8),
    relab9 = as.numeric(relab == 9)
  )

#Creo 5 variables binarias de tamaño de la firma
datos_filtrados2 <- datos_filtrados2 %>%
  mutate(
    sizeFirm1 = as.numeric(sizeFirm == 1),
    sizeFirm2 = as.numeric(sizeFirm == 2),
    sizeFirm3 = as.numeric(sizeFirm == 3),
    sizeFirm4 = as.numeric(sizeFirm == 4),
    sizeFirm5 = as.numeric(sizeFirm == 5)
  )


#Creo 12 variables binarias de tamaño de Mes
datos_filtrados2 <- datos_filtrados2 %>%
  mutate(
    mes1 = as.numeric(mes == 1),
    mes2 = as.numeric(mes == 2),
    mes3 = as.numeric(mes == 3),
    mes4 = as.numeric(mes == 4),
    mes5 = as.numeric(mes == 5),
    mes6 = as.numeric(mes == 6),
    mes7 = as.numeric(mes == 7),
    mes8= as.numeric(mes == 8),
    mes9 = as.numeric(mes == 9),
    mes10 = as.numeric(mes == 10),
    mes11 = as.numeric(mes == 11),
    mes12 = as.numeric(mes == 12)
  )


install.packages("caret")
install.packages("lattice")
install.packages("glmnet")
library(caret)
library(glmnet)

ruta_excel <- "archivo_filtrado3.xlsx"

# Leer el archivo Excel
datos_filtrados2 <- read_excel(ruta_excel)



# Planto la semilla
set.seed(123)


#Divido la muestra en entrenamiento y prueba
splitIndex <- createDataPartition(datos_filtrados2$log_y_salary_m_hu, p = 0.7, list = FALSE)
training_set <- datos_filtrados2[splitIndex, ]
testing_set <- datos_filtrados2[-splitIndex, ]

RMSE_valores <- numeric()

# Regresión lineal 1
lm1_model <- lm(log_y_salary_m_hu ~ 1, data = training_set)
lm_pred <- predict(lm1_model, newdata = testing_set)
lm1_rmse <- sqrt(mean((lm_pred - testing_set$log_y_salary_m_hu)^2))
RMSE_valores <- c(RMSE_valores, lm1_rmse)

# Regresión lineal 2
lm2_model <- lm(log_y_salary_m_hu ~ age , data = training_set)
lm_pred <- predict(lm2_model, newdata = testing_set)
lm2_rmse <- sqrt(mean((lm_pred - testing_set$log_y_salary_m_hu)^2))
RMSE_valores <- c(RMSE_valores, lm2_rmse)

# Regresión lineal 3
lm3_model <- lm(log_y_salary_m_hu ~ age + sex, data = training_set)
lm_pred <- predict(lm3_model, newdata = testing_set)
lm3_rmse <- sqrt(mean((lm_pred - testing_set$log_y_salary_m_hu)^2))
RMSE_valores <- c(RMSE_valores, lm3_rmse)

# Regresión lineal 4
lm4_model <- lm(log_y_salary_m_hu ~ age + sex + age_cuadrado, data = training_set)
lm_pred <- predict(lm4_model, newdata = testing_set)
lm4_rmse <- sqrt(mean((lm_pred - testing_set$log_y_salary_m_hu)^2))
RMSE_valores <- c(RMSE_valores, lm4_rmse)


# Regresión lineal 5
lm5_model <- lm(log_y_salary_m_hu ~ age  + sex + age_cuadrado + educ3 + educ4 + educ5 + educ6 + educ7, data = training_set)
lm_pred <- predict(lm5_model, newdata = testing_set)
lm5_rmse <- sqrt(mean((lm_pred - testing_set$log_y_salary_m_hu)^2))
RMSE_valores <- c(RMSE_valores, lm5_rmse)

# Regresión lineal 6
lm6_model <- lm(log_y_salary_m_hu ~ age  + sex + age_cuadrado + educ3 + educ4 + educ5 + educ6 + educ7, data = training_set)
lm_pred <- predict(lm6_model, newdata = testing_set)
lm6_rmse <- sqrt(mean((lm_pred - testing_set$log_y_salary_m_hu)^2))
RMSE_valores <- c(RMSE_valores, lm6_rmse)


# Regresión lineal 7
lm7_model <- lm(log_y_salary_m_hu ~ age + sex + age_cuadrado + educ3 + educ4 + educ5 + educ6 + educ7 + sizeFirm2 + sizeFirm3 + sizeFirm4 + sizeFirm5, data = training_set)
lm_pred <- predict(lm7_model, newdata = testing_set)
lm7_rmse <- sqrt(mean((lm_pred - testing_set$log_y_salary_m_hu)^2))
RMSE_valores <- c(RMSE_valores, lm7_rmse)

# Regresión lineal 8
lm8_model <- lm(log_y_salary_m_hu ~ age + sex + age_cuadrado + educ3 + educ4 + educ5 + educ6 + educ7 + formal, data = training_set)
lm_pred <- predict(lm8_model, newdata = testing_set)
lm8_rmse <- sqrt(mean((lm_pred - testing_set$log_y_salary_m_hu)^2))
RMSE_valores <- c(RMSE_valores, lm8_rmse)


# Regresión lineal 9
lm9_model <- lm(log_y_salary_m_hu ~ age + sex + age_cuadrado + educ3 + educ4 + educ5 + educ6 + educ7 + formal + sizeFirm2 + sizeFirm3 + sizeFirm4 + sizeFirm5, data = training_set)
lm_pred <- predict(lm9_model, newdata = testing_set)
lm9_rmse <- sqrt(mean((lm_pred - testing_set$log_y_salary_m_hu)^2))
RMSE_valores <- c(RMSE_valores, lm9_rmse)

# Regresión lineal 10
lm10_model <- lm(log_y_salary_m_hu ~ age + sex + age_cuadrado + educ3 + educ4 + educ5 + educ6 + educ7 + formal + sizeFirm2 + sizeFirm3 + sizeFirm4 + sizeFirm5 + mes2 + mes3 + mes4 + mes5 + mes6 + mes7 + mes8 + mes9 + mes10 + mes11 + mes12, data = training_set)
lm_pred <- predict(lm10_model, newdata = testing_set)
lm10_rmse <- sqrt(mean((lm_pred - testing_set$log_y_salary_m_hu)^2))
RMSE_valores <- c(RMSE_valores, lm10_rmse)

#Veo coeficientes estimados de cada regresión
summary_table <- summary(lm1_model)
coefficients_table <- summary_table$coefficients
print(coefficients_table)

summary_table <- summary(lm2_model)
coefficients_table <- summary_table$coefficients
print(coefficients_table)

summary_table <- summary(lm3_model)
coefficients_table <- summary_table$coefficients
print(coefficients_table)

summary_table <- summary(lm4_model)
coefficients_table <- summary_table$coefficients
print(coefficients_table)

summary_table <- summary(lm5_model)
coefficients_table <- summary_table$coefficients
print(coefficients_table)

summary_table <- summary(lm6_model)
coefficients_table <- summary_table$coefficients
print(coefficients_table)

summary_table <- summary(lm7_model)
coefficients_table <- summary_table$coefficients
print(coefficients_table)

summary_table <- summary(lm8_model)
coefficients_table <- summary_table$coefficients
print(coefficients_table)

summary_table <- summary(lm9_model)
coefficients_table <- summary_table$coefficients
print(coefficients_table)

summary_table <- summary(lm10_model)
coefficients_table <- summary_table$coefficients
print(coefficients_table)


cat("RMSE Values:\n")
cat(paste(names(RMSE_valores), ": ", round(RMSE_valores, 4), "\n"))


barplot(RMSE_valores, names.arg = names(RMSE_valores), col = "skyblue",
        main = "Comparación de RMSE", ylab = "RMSE")

# Crear el diagrama de barras

barplot(RMSE_valores, names.arg = names(RMSE_valores), col = "skyblue",
        main = "Comparación de RMSE", ylab = "RMSE", xlab = "Modelos", ylim = c(0, 1))

# Agregar etiquetas de datos a las barras
text(x = barplot(RMSE_valores, col = "skyblue", main = "", ylab = "RMSE", xlab = "Modelos", ylim = c(0, 1)),
     y = RMSE_valores, label = round(RMSE_valores, 3), pos = 3, cex = 0.8)

#Regresión 9 es la que tiene el menor RMSE
lm_model <- lm(log_y_salary_m_hu ~ age + sex + age_cuadrado + educ3 + educ4 + educ5 + educ6 + educ7 + formal + sizeFirm2 + sizeFirm3 + sizeFirm4 + sizeFirm5, data = training_set)
lm_pred <- predict(lm_model, newdata = testing_set)
lm_rmse <- sqrt(mean((lm_pred - testing_set$log_y_salary_m_hu)^2))

# Calcula errores de predicción para el modelo con menor RMSE
lm_errors <- testing_set$log_y_salary_m_hu - lm_pred


summary(lm_errors)

#Histograma de errores de predicción
hist(lm_errors, main = "Distribución de errores predichos", xlab = "Errores predichos")

# Identifica outliers
outlier_threshold <- 2 * IQR(lm_errors)
potential_outliers <- which(abs(lm_errors) > outlier_threshold)


#LOOCV
loocv_rmse_values <- numeric()

#Regresion lineal con LOOCV
lm_loocv_model <- train(log_y_salary_m_hu ~ age + sex + age_cuadrado + educ3 + educ4 + educ5 + educ6 + educ7 + formal + sizeFirm2 + sizeFirm3 + sizeFirm4 + sizeFirm5, data = datos_filtrados2, method = "lm",
                        trControl = trainControl(method = "LOOCV"))
lm_loocv_rmse <- lm_loocv_model$results$RMSE
loocv_rmse_values <- c(loocv_rmse_values, lm_loocv_rmse)

cat("\nComparison of Test Error and LOOCV Error:\n")
cat("Linear Regression - Test RMSE:", round(lm9_rmse, 4), ", LOOCV RMSE:", round(lm_loocv_rmse, 4), "\n")
```
